# ใช้ Matplotlib แสดงกราฟ

```python
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager, rcParams, rcParamsDefault
import pandas as pd
import scipy.stats as stats
from scipy.optimize import curve_fit
import numpy as np

font_manager.fontManager.addfont('thsarabunnew.ttf')
rc('font', family=['TH Sarabun New', 'DejaVu Sans'], size=14)
```

```python
data1_1 = pd.read_csv('./lab2_data/L3_data1-1.csv')
data1_2 = pd.read_csv('./lab2_data/L3_data1-2.csv')
# data2 = pd.read_csv('./lab2_data/L3_data2.csv')
```

# ข้อที่ 1

```python
x1 = data1_1['RL [Ohm]']
y1 = data1_1['V(RL) [V]']
x2 = data1_2['RL [Ohm]']
y2 = data1_2['V(RL) [V]']

# หา best fit curve
p1 = np.poly1d(np.polyfit(x1, y1, 7))
p2 = np.poly1d(np.polyfit(x2, y2, 7))

# วาดกราฟ
plt.xlabel(u'RL [\u03a9]')
plt.ylabel('V(RL) [V]')
plt.title(u'\u0e01\u0e23\u0e32\u0e1f\u0e41\u0e2a\u0e14\u0e07\u0e04\u0e27\u0e32\u0e21\u0e2a\u0e31\u0e21\u0e1e\u0e31\u0e19\u0e18\u0e4c\u0e23\u0e30\u0e2b\u0e27\u0e48\u0e32\u0e07\u0020\u0056\u0028\u0052\u004c\u0029\u0020\u0e01\u0e31\u0e1a\u0020\u0052\u004c')
plt.plot(x1, y1, 'o', label=u'\u0e15\u0e32\u0e23\u0e32\u0e07\u0e17\u0e35\u0e48\u0020\u0031')
plt.plot(x1, p1(x1), ls='--', label=u'\u0e15\u0e32\u0e23\u0e32\u0e07\u0e17\u0e35\u0e48\u0020\u0031')
plt.plot(x2, y2, 'o', label=u'\u0e15\u0e32\u0e23\u0e32\u0e07\u0e17\u0e35\u0e48\u0020\u0032')
plt.plot(x2, p2(x2), ls='--', label=u'\u0e15\u0e32\u0e23\u0e32\u0e07\u0e17\u0e35\u0e48\u0020\u0032')
plt.legend()
plt.savefig('./lab2_data/L3_data1.png')
plt.show()
```
