# ใช้ Matplotlib แสดงกราฟ

```python
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager
import pandas as pd
import scipy.stats as stats
from scipy.optimize import curve_fit
import numpy as np

font_manager.fontManager.addfont('thsarabunnew.ttf')
rc('font', family=['TH Sarabun New', 'DejaVu Sans'], size=14)
```

```python
data1 = pd.read_csv('./lab_data/L2_data1.csv')
data2 = pd.read_csv('./lab_data/L2_data2.csv')
```

# ข้อที่ 1

### กราฟความสัมพันธ์ระหว่าง V(R2) กับ I(R2) ตามกฎของโอห์ม

```python
# หา line of best fit
result = stats.linregress(data1['I(R2) [mA]'], data1['V(R2) [V]'])
y = result.slope * data1['I(R2) [mA]'] + result.intercept
```

```python
plt.scatter(data1['I(R2) [mA]'], data1['V(R2) [V]'])
plt.plot(data1['I(R2) [mA]'], y, '--', color='r')
plt.xlabel('I(R2) [mA]')
plt.ylabel('V(R2) [V]')
plt.title(u'\u0e01\u0e23\u0e32\u0e1f\u0e04\u0e27\u0e32\u0e21\u0e2a\u0e31\u0e21\u0e1e\u0e31\u0e19\u0e18\u0e4c\u0e23\u0e30\u0e2b\u0e27\u0e48\u0e32\u0e07\u0020\u0056\u0028\u0052\u0032\u0029\u0020\u0e01\u0e31\u0e1a\u0020\u0049\u0028\u0052\u0032\u0029\u0020\u0e15\u0e32\u0e21\u0e01\u0e0e\u0e02\u0e2d\u0e07\u0e42\u0e2d\u0e2b\u0e4c\u0e21')
plt.savefig('./lab2_data/L2_data1-1.png')
plt.show()
```

# ข้อที่ 2

### กราฟความสัมพันธ์ระหว่าง P(R2) กับ R2

```python
def objective(x, a, b, c, d, e, f):
    return (a * x) + (b * x**2) + (c * x**3) + (d * x**4) + (e * x**5) + f

# หา best fit curve
popt, _ = curve_fit(objective, data2['R2 [Ohm]'], data2['P(R2) [mW]'])
a, b, c, d, e, f = popt
x_line = np.arange(min(data2['R2 [Ohm]']), max(data2['R2 [Ohm]']), 1)
y_line = objective(x_line, a, b, c, d, e, f)
```

```python
plt.scatter(data2['R2 [Ohm]'], data2['P(R2) [mW]'])
plt.plot(x_line, y_line, '--', color='r')
plt.xlabel(u'R2 [\u2126]')
plt.ylabel('P(R2) [mW]')
plt.title(u'\u0e01\u0e23\u0e32\u0e1f\u0e04\u0e27\u0e32\u0e21\u0e2a\u0e31\u0e21\u0e1e\u0e31\u0e19\u0e18\u0e4c\u0e23\u0e30\u0e2b\u0e27\u0e48\u0e32\u0e07\u0020\u0050\u0028\u0052\u0032\u0029\u0020\u0e01\u0e31\u0e1a\u0020\u0052\u0032')
plt.savefig('./lab2_data/L2_data2.png')
plt.show()
```
