# ใช้ Matplotlib แสดงกราฟ

```python
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager, rcParams, rcParamsDefault
import pandas as pd
import scipy.stats as stats
from scipy.optimize import curve_fit
import numpy as np

font_manager.fontManager.addfont('thsarabunnew.ttf')
rc('font', family=['TH Sarabun New', 'DejaVu Sans'], size=14)
```

```python
data1 = pd.read_csv('./lab_data/L5_data1.csv')
data2 = pd.read_csv('./lab_data/L5_data2.csv')
data3 = pd.read_csv('./lab_data/L5_data3.csv')
```

```python
def make_graph(data, title, save_name):
    x = data.iloc[1:]['X'].astype(float) * data.iloc[0]['Increment'].astype(float)
    y = data.iloc[1:]['CH1'].astype(float)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    ax1.grid()
    ax1.plot(x, y, label='CH1')
    ax1.set_title('CH1')
    ax2.grid()
    y = data.iloc[1:]['CH2'].astype(float)
    ax2.plot(x, y, label='CH2')
    ax2.set_title('CH2')
    fig.supxlabel('Time (s)')
    fig.supylabel('Voltage (V)')
    fig.suptitle(title)
    fig.savefig('./lab_data/' + save_name)
    plt.show()
```

```python
make_graph(data1, 'Data 1', 'L5_fig1.png')
```

```python
make_graph(data2, 'Data 2', 'L5_fig2.png')
```

```python
make_graph(data3, 'Data 3', 'L5_fig3.png')
```
