```python
from sympy import solve, symbols, sin, pi, plot, Eq
```

```python
x = symbols('x')
f = 2.5 * sin(100 * pi * x)
g = 2.5 * sin(100 * pi * x + pi / 2)
y = f + g
Eq(symbols('y'), y)
```

```python
plot(f, g, y, (x, 0, 16*2e-3), size=(15,5), xlabel='Time (s)', ylabel='Voltage (V)', legend=True)
```
