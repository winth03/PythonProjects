```python
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d

font_manager.fontManager.addfont('thsarabunnew.ttf')
rc('font', family=['TH Sarabun New', 'DejaVu Sans'], size=14)
```

```python
def unit_convert(elem):
    elem = str(elem)
    if "m" in elem:
        return float(elem[:-1]) * 1e-3
    elif "Î¼" in elem:
        return float(elem[:-1]) * 1e-6
    return float(elem)
```

```python
data1 = pd.read_csv("./lab_data/L11_data1.csv").applymap(unit_convert)
data2 = pd.read_csv("./lab_data/L11_data2.csv").applymap(unit_convert)
data3 = pd.read_csv("./lab_data/L11_data3.csv").applymap(unit_convert)
```

```python
x = data1["VGS [V]"]
y = data1["ID [A]"]
plt.plot(x, y)
plt.title("1.2 - 1")
plt.xlabel("VGS [V]")
plt.ylabel("ID [A]")
```

```python
y = data1["VDS [V]"]
plt.plot(x, y)
plt.title("1.2 - 2")
plt.xlabel("VGS [V]")
plt.ylabel("VDS [V]")
```
